```python
import pandas as pd

# Get all unique columns
all_columns = df_other.columns.union(df_zimbabwe.columns)

# Align columns, keeping missing ones as NaN
df_other = df_other.reindex(columns=all_columns)
df_zimbabwe = df_zimbabwe.reindex(columns=all_columns)

# Stack them together (ignore index for clean stacking)
df_combined = pd.concat([df_other, df_zimbabwe], ignore_index=True)

# Save the combined dataframe if needed
folder_path = "Files"  # Replace with the actual path if different

# Create the folder if it doesn't exist
if not os.path.exists(folder_path):
    os.makedirs(folder_path)

# Save the CSV file in the "Files" folder
df_combined.to_csv(os.path.join(folder_path, "combined_output.csv"), index=False)


# Print preview and check for row counts
df_combined

if len(df_zimbabwe)+len(df_other) == len(df_combined):
    print('Same number of rows for combined dataset')
else:
    print('double check')
```
